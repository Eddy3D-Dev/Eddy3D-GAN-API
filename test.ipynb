```python
# Open your image file in binary mode
with open("test_input.png", "rb") as f:
    img_binary = f.read()


# Create a file-like object from the image binary data
img_file = io.BytesIO(img_binary)

# Open the image using PIL
img_in = Image.open(img_file)

# Display the image
img_in.show()


# Send the POST request
# response = requests.post('http://127.0.0.1:8000/process_image', files={'file': img_binary})
response = requests.post(
    "https://gan-230522.eddy3d.com/process_image", files={"file": img_binary}
)


# The response is an image file, so we need to convert it back to an image
image = Image.open(io.BytesIO(response.content))

# Now you can display the image or save it to a file
image.show()
```
